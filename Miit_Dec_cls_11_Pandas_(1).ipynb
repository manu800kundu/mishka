```python
import seaborn as sns
df=sns.load_dataset('titanic')
df
```

```python
df['who'].value_counts()
```

```python
df['sex'].value_counts()
```

```python
# drop all the unwanted cols
df.drop(['sex', 'alone', 'embarked', 'pclass', 'survived', 'adult_male'], axis=1, inplace=True)
```

```python
# check weather the remaining dataset is cleaned or not
df.isna().sum().sort_values(ascending=False)/len(df) *100
```

```python
df.drop('deck', axis=1, inplace=True)
```

```python
# check weather the remaining dataset is cleaned or not
df.isna().sum().sort_values(ascending=False)/len(df) *100
```

```python
# whenever you have a numerical col which have missing values and you can fill it
# with either mean or median of that col
df['age']=df['age'].fillna(df['age'].median())

# whenever you have a cat col which have missing values and you can fill it
# with mode of that col
df['embark_town']=df['embark_town'].fillna(df['embark_town'].mode()[0])
```

```python
# check weather the remaining dataset is cleaned or not
df.isna().sum().sort_values(ascending=False)/len(df) *100
```

```python
df.duplicated().sum()
```

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```
